```python
class agenda:
    
    def __init__(self):
        self.contactos={}
        
    def menu(self):
        posicion=0
        while posicion!="5":
            print("1- Cargar contacto en agenda")
            print("2- Listar contactos de la agenda")
            print("3- Consultar contactos de la agenda.")
            print("4- Modificación de teléfono y mail.")
            print("5- Salir del menú.")
            print("------------------")
            posicion=input("Elija una opción del menú: ")
            if posicion=="1":
                self.cargar()
            if posicion=="2":
                self.listar()
            if posicion=="3":
                self.consultar()
            if posicion=="4":
                self.modificar()
        
            
    def cargar(self):
        print("------------------")
        print("Opción 1. CARGAR CONTACTO EN AGENDA")
        self.cantidad=int(input("¿Cuántos contactos quiere guardar?: "))
        for x in range(self.cantidad):
            nombre=input("Ingrese el nombre del contacto: ")
            telefono=input("Ingrese el teléfono del contacto: ")
            mail=input("Ingrese el mail del contacto: ")
            self.contactos[nombre]=[telefono, mail]
        print("------------------")

                   
    def listar(self):
        print("------------------")
        print("Opción 2. LISTAR CONTACTOS DE LA AGENDA.")
        for nombre in self.contactos:
            print(nombre, self.contactos[nombre][0], self.contactos[nombre][1])
        print("------------------")
        print(self.contactos)
        print("------------------")
            
        
    def consultar(self):
        print("------------------")
        print("Opción 3. CONSULTAR CONTACTOS DE LA AGENDA.")
        nombre=input("Ingrese el nombre del contacto a consultar: ")
        if nombre in self.contactos:
            print("Los datos del contacto: ", nombre, "son", self.contactos[nombre][0],"/", self.contactos[nombre][1])
        else:
            print("ERROR: No existe ningún contacto con ese nombre.")
        print("------------------")
           
    def modificar(self):
        print("------------------")
        print("Opción 3. MODIFICACIÓN DE TELÉFONO Y MAIL.")        
        nombre=input("Ingrese el nombre del contacto a modificar: ")
        if nombre in self.contactos:
            print("El contacto elegido para modificar es: ", nombre)
            print("Ingrese '1' para modificar solo el teléfono.\n Ingrese '2' para modificar solo el mail.\n Ingrese '3' para modificar el teléfono y el mail:")
            posicion=input("-->")
            if posicion=="1":
                if nombre in self.contactos:
                    nuevotelefono=input("Ingrese el nuevo teléfono del contacto: ")
                    self.contactos[nombre][0]=nuevotelefono
                    print("Teléfono del contacto modificado")
                else:
                    print("ERROR: No existe ningún contacto con ese nombre.")
            if posicion=="2":
                if nombre in self.contactos:
                    nuevomail=input("Ingrese el nuevo mail del contacto: ")
                    self.contactos[nombre][1]=nuevomail
                    print("Mail del contacto modificado.")
                else:
                    print("ERROR: No existe ningún contacto con ese nombre.")
            if posicion=="3":
                if nombre in self.contactos:
                    nuevotelefono=input("Ingrese el nuevo teléfono del contacto: ")
                    self.contactos[nombre][0]=nuevotelefono
                    print("Teléfono del contacto modificado.")
                    nuevomail=input("Ingrese el nuevo mail del contacto: ")
                    self.contactos[nombre][1]=nuevomail
                    print("Mail del contaco modificado.")
                else:
                    print("ERROR: No existe ningún contacto con ese nombre.")
            if posicion!="1" and posicion!="2" and posicion!="3":
                print("Opcion no válida. Saliendo del menú de modificación de contacto.")


#diccionario para no tener que introducir todo el rato los datos en las comprobaciones.
#dic={"Pedro":["698236541", "pedro@hotmail.com"], "Ana":["698720136", "ana@hotmail.com"], "Luis":["602354820", "luis@hotmail.com"]}
agenda=agenda()
agenda.menu()
```
